```python
# 마운트 및 csv파일 출력

import pandas as pd
import os

# 디렉토리 경로와 파일명을 결합하여 파일 시스템 내 절대 경로를 확정
file_path = os.path.join(folder_dir, DATA_FILE)

# CSV 데이터 파싱 및 메모리 내 DataFrame 객체 생성
try:
    df = pd.read_csv(file_path)
    
    # Colab 인터랙티브 환경의 데이터 렌더링 기능을 활용한 출력
    display(df)
except FileNotFoundError:
    print(f"오류: 지정된 경로에서 파일을 식별할 수 없습니다. ({file_path})")
```

```python
# 테스트셋, 학습셋 형성
X = df.drop('Outcome',axis = 1)
X = X.drop('BMI',axis = 1)
display(X)

Y=df['BMI']
display(Y)

## 테스트 집단 추출

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR

import pandas as pd
import numpy as np

```

```python
diabeteLinearRegression = LinearRegression()
diabeteLinearRegression.fit(X_train, Y_train)
predLinearRegression = diabeteLinearRegression.predict(X_test)

print('평균제곱근오차', mean_squared_error(predLinearRegression, Y_test))
```

```python
diabeteDecisionTreeRegressor = DecisionTreeRegressor()
diabeteDecisionTreeRegressor.fit(X_train, Y_train)
predDecisionTreeRegressor = diabeteDecisionTreeRegressor.predict(X_test)

print('평균제곱근오차', mean_squared_error(predDecisionTreeRegressor, Y_test))
```

```python
diabeteRandomForestRegressor = RandomForestRegressor()
diabeteRandomForestRegressor.fit(X_train, Y_train)
predRandomForestRegressor = diabeteRandomForestRegressor.predict(X_test)

print('평균제곱근오차', mean_squared_error(predRandomForestRegressor, Y_test))
```

```python
diabeteSVR = SVR()
diabeteSVR.fit(X_train, Y_train)
predSVR = diabeteSVR.predict(X_test)

print('평균제곱근오차', mean_squared_error(predSVR, Y_test))
```
