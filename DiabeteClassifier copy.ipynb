```python
# 마운트 및 csv파일 출력

import pandas as pd
import os

# 디렉토리 경로와 파일명을 결합하여 파일 시스템 내 절대 경로를 확정
file_path = os.path.join(folder_dir, DATA_FILE)

# CSV 데이터 파싱 및 메모리 내 DataFrame 객체 생성
try:
    df = pd.read_csv(file_path)
    
    # Colab 인터랙티브 환경의 데이터 렌더링 기능을 활용한 출력
    display(df)
except FileNotFoundError:
    print(f"오류: 지정된 경로에서 파일을 식별할 수 없습니다. ({file_path})")
```

```python
# 테스트셋, 학습셋 형성
X = df.drop('Outcome',axis = 1)
display(X)

Y=df['Outcome']
display(Y)

## 테스트 집단 추출

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y,test_size=0.25,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

import warnings
warnings.filterwarnings('ignore')
```

```python
diabeteLogisticRegression = LogisticRegression(random_state=0)
diabeteLogisticRegression.fit(X_train, Y_train)

predLogisticRegression = diabeteLogisticRegression.predict(X_test)

print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(Y_test, predLogisticRegression))
print (confusion_matrix(Y_test, predLogisticRegression))
```

```python
diabeteMLPClassifier = MLPClassifier(random_state=0)
diabeteMLPClassifier.fit(X_train, Y_train)

predMLPClassifier = diabeteMLPClassifier.predict(X_test)

print ("\n--- Neural Network Classifier ---")
print (accuracy_score(Y_test, predMLPClassifier))
print (confusion_matrix(Y_test, predMLPClassifier))
```

```python
print ("\n--- Radom Forest ---")
diabeteRandomForestClassifier = RandomForestClassifier(random_state=0)
diabeteRandomForestClassifier.fit(X_train, Y_train)
predRandomForestClassifier = diabeteRandomForestClassifier.predict(X_test)
print(accuracy_score(Y_test,predRandomForestClassifier))
print (confusion_matrix(Y_test, predRandomForestClassifier))
```

```python
diabeteSVC = SVC(random_state=0)
diabeteSVC.fit(X_train, Y_train)

predSVC = diabeteSVC.predict(X_test)

print("\n--- SVM Classifier ---")
print(accuracy_score(Y_test, predSVC))
print(confusion_matrix(Y_test, predSVC))
```

```python

# Build a decision tree classifier and predict

diabeteDecisionTreeClassifier = DecisionTreeClassifier(random_state=0)
diabeteDecisionTreeClassifier.fit(X_train, Y_train)

predDecisionTreeClassifier = diabeteDecisionTreeClassifier.predict(X_test)

print ("\n--- Decision Tree Classifier ---")
print (accuracy_score(Y_test, predDecisionTreeClassifier))
print (confusion_matrix(Y_test, predDecisionTreeClassifier))
```
